```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = [4, 5, 9, 4, 3, 11, 14 , 6, 10, 12,22, 23, 21, 25]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 18,10, 15,16, 12]

plt.scatter(x, y)
plt.show()
```

```python
data = list(zip(x,y))
inertias = []
for i in range(1,15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,15), inertias ,marker='*' )
plt.title('elbow method')
plt.xlabel('clusters')
plt.ylabel('interia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
